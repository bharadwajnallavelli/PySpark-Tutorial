```
arr_holdings = [('user1',['aapl','msft','spy','pdd']),
                ('user2',['intc','nvda','spx','amzn']),
                ('user3',['aapl',None]),
                ('user4', None)]
arr_schema = ['username', 'stocks']
df_arr = spark.createDataFrame(data = arr_holdings, schema = arr_schema)
```

```
display(df_arr.printSchema())
display(df_arr)
```

```
map_holdings = [('user1',{'aapl':'2','msft':'1','spy':'2','pdd':'4'}),
                ('user2',{'intc':'2','nvda':'4','spx':'6','amzn':'8'}),
                ('user3',{'aapl':'5'}),
                ('user4', None)]
map_holdings_schema = ['username','stockdetails']
df_map = spark.createDataFrame(data = map_holdings, schema = map_holdings_schema)
```

```
display(df_map.printSchema())
display(df_map)
```

```
from pyspark.sql.functions import explode
df2 = df_arr.select(df_arr.username, explode(df_arr.stocks))
```

```
display(df2.printSchema())
display(df2)
```

```
df3 = df_map.select(df_map.username, explode(df_map.stockdetails))
```

```
display(df3.printSchema())
display(df3)
```

```
from pyspark.sql.functions import explode_outer
display(df_map.select(df_map.username, explode_outer(df_map.stockdetails)))
```

```
display(df_arr.select(df_arr.username, explode_outer(df_arr.stocks)))
```
