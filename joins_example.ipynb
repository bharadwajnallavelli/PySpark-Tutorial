```
employee_data = [(1,"Forbes","1987","100","Male",1680),
                (2,"Renado","2008","200","Male",1676),
                (3,"Emilee","2012","700","Female",1294),
                (4,"Chlo","2012","400","Female",1378),
                (5,"Jaye","2009","500","Male",1835)]
employee_data_schema = ["employee_id","name","doj","department_id","gender","salary"]
```

```
employee_df = spark.createDataFrame(data = employee_data, schema = employee_data_schema)
```

```
department_data = [(100, "HR"),(200, "Accounts"), (300, "IT"), (400, "Sales"), (500, "Delivery"), (600, "Comms")]
department_data_schema = ["department_id","department_name"]
```

```
department_df = spark.createDataFrame(data = department_data, schema = department_data_schema)
```

```
display(employee_df)
display(department_df)
```

#### Inner Join

```
inner_df = employee_df.join(department_df, employee_df.department_id == department_df.department_id, "inner")
```

```
display(inner_df)
```

#### left

```
left_df = employee_df.join(department_df, employee_df.department_id == department_df.department_id, "left")
display(left_df)
```

#### Right

```
right_df = employee_df.join(department_df, employee_df.department_id == department_df.department_id, "right")
display(right_df)
```

#### Outer

```
outer_df = employee_df.join(department_df, employee_df.department_id == department_df.department_id, "outer")
display(outer_df)
```

#### Left Semi

```
left_semi_df = employee_df.join(department_df, employee_df.department_id == department_df.department_id, "semi")
display(left_semi_df)
```

#### Anti

```
anti_df = employee_df.join(department_df, employee_df.department_id == department_df.department_id, "anti")
display(anti_df)
```
