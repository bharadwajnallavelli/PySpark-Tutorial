```
data = [("company1","Q1",2000),
       ("company1","Q2",2400),
       ("company1","Q3",1000),
       ("company1","Q4",3000),
       ("company2","Q1",2000),
       ("company2","Q2",1000),
       ("company2","Q3",8000),
       ("company2","Q4",9000),
       ("company3","Q1",2900),
       ("company3","Q2",200),
       ("company3","Q3",1000),
       ("company3","Q4",1500)]
data_schema= ['company','quarter','revenue']
```

```
df_data = spark.createDataFrame(data = data, schema = data_schema)
display(df_data.printSchema())
display(df_data)
```

```
pivot_df = df_data.groupBy('company').pivot('quarter').sum('revenue')
display(pivot_df.printSchema())
display(pivot_df)
```

```
DF = pivot_df.selectExpr("company", "stack(4,'Q1',Q1,'Q2',Q2,'Q3',Q3,'Q4',Q4) as (Quarter, Revenue)")
display(DF)
```
