```
data = [('student1','80','P','90'),
       ('student2','90','P','80'),
       ('student3','70','P','60'),
       ('student4','50','F','30'),
       ('student5', None , 'NA','10')]
data_schema = ['student_name','marks','status','attendance']
df = spark.createDataFrame(data = data, schema = data_schema)
display(df.printSchema())
display(df)
```

```
from pyspark.sql.functions import when
df1 = df.withColumn('Grade', when((df.marks >= 80) & (df.attendance > 80), "Distinction")
                    .when((df.marks >= 70) & (df.attendance >= 60), "Good")
                    .when((df.marks >= 50) & (df.attendance <= 50), "Bad")
                    .otherwise("Not Applicable"))
display(df1)
```

```
df1 = df1.withColumn('status', when(df.status == 'P', 'Pass')
                    .when(df.status == 'F', 'Fail')
                    .otherwise("Not Applicable"))
display(df1)
```
